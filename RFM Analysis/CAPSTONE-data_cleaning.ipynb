```python
order_data_csv = pd.read_csv(r'C:\Users\99617\Desktop\JUPYTER EXERCISE\CAPSTONE\capstone_dataset\orders_dataset.csv')
order_data_csv.head(1)
```

```python
prod_data_csv = pd.read_csv(r'C:\Users\99617\Desktop\JUPYTER EXERCISE\CAPSTONE\capstone_dataset\products_dataset.csv') ##okay na merge_4
prod_data_csv.head(1)
```

```python
prod_categ = pd.read_csv(r'C:\Users\99617\Desktop\JUPYTER EXERCISE\CAPSTONE\capstone_dataset\product_category_name_translation.csv') ##okay na merge_4
prod_categ.head(1)
```

```python
merge_4 = pd.merge(prod_data_csv, prod_categ,
                   left_on = 'product_category_name',
                   right_on = 'product_category_name',
                   how = 'inner')
merge_4.head(2)
```

```python
merge_4.drop(['product_name_lenght','product_description_lenght','product_photos_qty','product_weight_g','product_length_cm',
'product_height_cm','product_width_cm',], axis=1, inplace = True)
merge_4.head(1)
```

```python
merge_1 = pd.merge(order_csv, order_payment_csv,
                   left_on = 'order_id',
                   right_on = 'order_id',
                   how = 'inner')
merge_1.head(2)
```

```python
merge_2 = pd.merge(merge_1, merge_4,
                   left_on = 'product_id',
                   right_on = 'product_id',
                   how = 'inner')
merge_2.head(2)
```

```python
merge_3 = pd.merge(merge_2, order_data_csv,
                   left_on = 'order_id',
                   right_on = 'order_id',
                   how = 'inner')
merge_3.head(2)
```

```python
merge_final = pd.merge(cust_csv, merge_3,
                   left_on = 'customer_id',
                   right_on = 'customer_id',
                   how = 'inner')
merge_final.head(2)
```

```python
merge_final.info()
```

```python
merge_final.drop(['customer_zip_code_prefix','price','freight_value','product_category_name','seller_id','shipping_limit_date','order_estimated_delivery_date','order_approved_at'], axis=1, inplace = True)
merge_final.head(1)
```

```python
merge_final['customer_id'] = merge_final['customer_id'].str[-15:]
merge_final['order_id'] = merge_final['order_id'].str[-15:]
merge_final['product_id'] = merge_final['product_id'].str[-15:]
merge_final['customer_unique_id'] = merge_final['customer_unique_id'].str[-15:]

print("Number of unique in cust_id",merge_final['customer_id'].nunique())
print("Number of unique in order_id",merge_final['order_id'].nunique())
print("Number of unique in prod_id",merge_final['product_id'].nunique())
print("Number of unique in cust_un_id",merge_final['customer_unique_id'].nunique())
```
