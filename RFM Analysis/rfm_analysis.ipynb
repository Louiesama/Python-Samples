## RFM

```python
analysis_date = pd.Timestamp('2018-09-30')

rfm = df.groupby('customer_unique_id').agg({
    'order_purchase_timestamp':lambda x: (analysis_date - x.max()).days,
    'order_id': 'count',
    'payment_value': 'sum'})
rfm.columns = ['Recency','Frequency','Monetary']
rfm = rfm.reset_index()

df = df.merge(rfm, on='customer_unique_id', how='left')
df.head()
```

## SCORING SYSTEM

```python
df['R_score']=pd.cut(
    df['Recency'],
    bins=[1,60,120,360,600, float('inf')],
    labels=[5,4,3,2,1]
)

df['F_score']=pd.cut(
    df['Frequency'],
    bins=[0,10,15,20,30, float('inf')],
    labels=[1,2,3,4,5]
)

df['M_score']=pd.cut(
    df['Monetary'],
    bins=[1,50,100,200,500, float('inf')],
    labels=[1,2,3,4,5]
)

def segment_cust(row):
    r = row['R_score']
    f = row['F_score']
    m = row['M_score']

    if r >= 4 and f >= 4 and m >= 4:
        return 'Champions'
        
    if f >= 4 and r >= 3:
        return 'Loyal Customers'

    if r <= 2 and (f >= 3 or m >= 3):
        return 'At Risk'
        
    if r >= 4 and f <= 2:
        return 'New Customers'

    if r <= 2 and f <= 2 and m <= 2:
        return 'Loss'

    if m >= 4 and f <= 2:
        return 'Big Spenders'

    else:
        return 'Need Attention'

df['Segment'] = df.apply(segment_cust, axis=1)

df['Segment'].value_counts()
```

## DATA VIZ/DASHBBOARDS

```python
fig, axes = plt.subplots(2,3, figsize=(20,12))

## Count of segments
axes[0,0]=sns.countplot(data=df,x='Segment', order=df['Segment'].value_counts().index, 
                        palette = 'viridis', ax=axes[0,0],edgecolor='black')

for container in axes[0,0].containers:
    axes[0,0].bar_label(container)
    
axes[0,0].set_title('Volume of Segmentation', fontweight='bold', fontsize=15)
axes[0,0].set_xlabel('Segment type')
axes[0,0].set_ylabel('Count')
axes[0,0].tick_params(axis='x', rotation=45)


## Sales per Segments
sum_total = df.groupby('Segment')['Monetary'].sum().sort_values(ascending = False).index
axes[0,1]=sns.barplot(data=df, x='Segment', y='Monetary', order=sum_total, 
                      palette = 'viridis', ax=axes[0,1], errorbar=None,  edgecolor='black')
for container in axes[0,1].containers:
    axes[0,1].bar_label(container)

axes[0,1].set_title('Profit per Segmentation', fontweight='bold', fontsize=15)
axes[0,1].set_xlabel('Segment type')
axes[0,1].set_ylabel('Sum')
axes[0,1].tick_params(axis='x', rotation=45)


## Correlation between RFM
corr_matrix = df[['Recency','Frequency','Monetary']].corr()
axes[0,2]=sns.heatmap(corr_matrix, annot = True, cmap='coolwarm', fmt=".2f", linewidth=.5, cbar=True, ax=axes[0,2], edgecolor='black')
axes[0,2].set_title('RFM Variables Correlation Map', fontweight='bold', fontsize=15)
axes[0,2].tick_params(axis='x', rotation=45)


## Profit per R_score
r_scores = df.groupby('R_score')['Monetary'].sum().sort_values(ascending=False).index
axes[1,0]=sns.barplot(data=df, x='R_score',y='Monetary', order=r_scores, 
                      palette = 'viridis', ax=axes[1,0], errorbar=None, edgecolor='black')
for container in axes[1,0].containers:
    axes[1,0].bar_label(container)

axes[1,0].set_title('Profit per R_score', fontweight='bold', fontsize=15)
axes[1,0].set_xlabel('Higher number = Recent Buyer')
axes[1,0].set_ylabel('Amount')


## Profit per F_score
f_scores = df.groupby('F_score')['Monetary'].sum().sort_values(ascending=False).index
axes[1,1]=sns.barplot(data=df, x='F_score',y='Monetary', order=f_scores, 
                      palette = 'viridis', ax=axes[1,1], errorbar=None, edgecolor='black')
for container in axes[1,1].containers:
    axes[1,1].bar_label(container)

axes[1,1].set_title('Profit per F_score', fontweight='bold', fontsize=15)
axes[1,1].set_xlabel('Higher number = Frequent Buyer')
axes[1,1].set_ylabel('Amount')


## Profit per M_score
m_scores = df.groupby('M_score')['Monetary'].sum().sort_values(ascending=False).index
axes[1,2]=sns.barplot(data=df, x='M_score',y='Monetary', order=r_scores, 
                      palette = 'viridis', ax=axes[1,2], errorbar=None, edgecolor='black')
for container in axes[1,2].containers:
    axes[1,2].bar_label(container)

axes[1,2].set_title('Profit per M_score', fontweight='bold', fontsize=15)
axes[1,2].set_xlabel('Higher number = Big Spender')
axes[1,2].set_ylabel('Amount')

plt.suptitle('RFM Analysis Dashboard', fontsize=17, fontweight = 'bold', y=1.02)
plt.tight_layout()
plt.show()
```
