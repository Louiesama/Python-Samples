**Analyze**

```python
# GUIDE
## 1. Calculate Q1, Q3, and IQR for each region
## 2. Identify outliers for each region (values beyond Q3 + 1.5*IQR)
## 3. Calculate coefficient of variation (cv formula: std/mean) for each region → Lower CV = more consistent
## 4. Compare regions: Which is most consistent? Most variable?

# Your Code:
## 1. Calculate Q1, Q3, and IQR for each region

quartile = store_csv.groupby('Region')['Sales'].quantile([0.25,0.75]).unstack()
quartile.columns = ['Q1','Q3']

quartile['IQR'] = quartile['Q3'] - quartile['Q1']

print(quartile)
```

```python
## 2. Identify outliers for each region (values beyond Q3 + 1.5*IQR)

quartile['Upper_Bound'] = quartile['Q3'] + 1.5 * quartile['IQR']
quartile['Lower_Bound'] = quartile['Q1'] - 1.5 * quartile['IQR']

df_bounds = store_csv.merge(quartile[['Lower_Bound','Upper_Bound']], left_on = 'Region', right_index=True)
outliers = df_bounds[(df_bounds['Sales'] > df_bounds['Upper_Bound']) | (df_bounds['Sales'] < df_bounds['Lower_Bound'])]
print(outliers[['Region', 'Sales', 'Lower_Bound', 'Upper_Bound']].sort_values(by='Region'))
```

```python
## 3. Calculate coefficient of variation (cv formula: std/mean) for each region → Lower CV = more consistent

cv_data = store_csv.groupby('Region')['Sales'].agg(['mean', 'std'])
cv_data.columns = ['Mean', 'Std']

cv_data['CV'] = cv_data['Std'] / cv_data['Mean']
cv_data_sort = cv_data.sort_values(by='CV', ascending=True)
print(cv_data_sort)
```

```python
## 4. Compare regions: Which is most consistent? Most variable?
print(f" Region most consistent is: {cv_data_sort.head(1)}")
print(f" Region most variable is: {cv_data.sort_values(by='CV', ascending=False).head(1)}")
```

```python
# GUIDE
## 1. Create a histogram of Profit with 30 bins (Observe: is it symmetrical? or skewed?)
## 2. Calculate: - Percentage of orders with negative profit (loss percentage) and - Average profit for profitable vs unprofitable orders
## 3. Create 2 histograms side by side: -All orders and - Only profitable orders

# Your Code:

## 1. Create a histogram of Profit with 30 bins (Observe: is it symmetrical? or skewed?)

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(store_csv['Profit'], bins=30, range=(0, 30), color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Histogram by Profit', fontsize=15, fontweight='bold')
plt.xlabel('Profit', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.xticks(range(0, 31, 2))
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
## 2. Calculate: - Percentage of orders with negative profit (loss percentage) and - Average profit for profitable vs unprofitable orders

total_orders = len(store_csv)
unprofit = store_csv[store_csv['Profit'] < 0]
num_unprofit = len(unprofit)

loss_percent = (num_unprofit / total_orders) * 100

profit_orders = store_csv[store_csv['Profit'] >= 0]
avg_prof = profit_orders['Profit'].mean()

avg_unprof = unprofit['Profit'].mean()

results_df = pd.DataFrame({'Metric': ['Percentage of Orders with Negative Profit (Loss Percentage)',
                                      'Average Profit for Profitable Orders',
                                      'Average Profit for Unprofitable Orders (Loss)'],
                           'Value': [f'{loss_percent:.2f}%',
                                     f'${avg_prof:,.2f}',
                                     f'${avg_unprof:,.2f}']})

print(results_df)
```
